```python
# 商品名を格納するリストを作成します
product_names = []
for i in range(1,5):
    # URLの指定
    urlName = "https://seijoishii.com/collections/cat01?page="+str(i)

    # ページの取得
    time.sleep(3)
    page = requests.get(urlName)
    page.encoding = page.apparent_encoding
    #pprint(page.content)

    # BeautifulSoupオブジェクトの生成
    soup = BeautifulSoup(page.content, 'html.parser')
    # pprint(soup)
    
    # 商品名をすべて取得してリストに格納します
    for product in soup.find_all('p', {'class': 'card__heading hb-cardProduct_title h5'}):
        product_name = product.text.strip()
        product_names.append(product_name)

```

# 全体

```python
from tqdm import tqdm  # Import tqdm

item_names = []
category = []
cate = ["デザート・スイーツ・パン",
        "お菓子",
        "お惣菜",
        "生鮮食品・米",
        "乳製品・加工肉・オードブル",
        "漬物・麺類・和食材",
        "調味料・レトルト・乾物",
        "コーヒー・お茶",
        "ジャム・蜂蜜・シリアル",
        "冷凍食品",
        "ドリンク・水",
        "ワイン・お酒",
        "サービス・雑貨",
        "ギフト・贈り物",
        "成城石井自家製",
        "業務用・大容量"]

for i in tqdm(range(1, 209), desc="Fetching Pages"):  # Wrap the loop with tqdm
    # URLの指定
    urlName = "https://seijoishii.com/collections/all?page=" + str(i)

    # ページの取得
    time.sleep(3)
    page = requests.get(urlName)
    page.encoding = page.apparent_encoding

    # BeautifulSoupオブジェクトの生成
    soup = BeautifulSoup(page.content, 'html.parser')

    # 各商品を含む<button>タグを取得します
    buttons = soup.find_all('button', class_='hb-cardProduct_cartBtn')

    # 各<button>タグから商品名と商品タグを抽出して辞書に格納します
    for button in buttons:
        product_title = button.get('data-product-title')
        product_tags = button.get('data-product-tags').split(',')
        for j in range(len(cate)):
            if cate[j] in product_tags:
                item_names.append(product_title)
                category.append(cate[j])
                break
```

```python
import pandas as pd
# リストをデータフレームの列に変換
df = pd.DataFrame({'item': item_names, 'category': category})

display(df)
```

```python
# データフレームをCSVファイルに保存
#folder_path = 'review'
#df.to_excel('review/伏見稲荷大社口コミ.csv', index=False)
import csv
file_name = 'item_data.csv'

# データフレームをCSVファイルに保存
with open(file_name, mode='w', newline='') as file:
    writer = csv.writer(file)
    # ヘッダーを書き込む
    writer.writerow(df.columns)
    # 行を書き込む
    for row in df.values:
        writer.writerow(row)
```

```python
item_names = []
label = []
category = []
for i in tqdm(range(1, 201), desc="Fetching Pages"):
    # URLの指定
    if i == 1:
        urlName = "https://ns.yaoko-net.com/products?sort=recommend&with_filter="
    else:
        urlName = "https://ns.yaoko-net.com/products?page="+str(i)+"&sort=recommend&with_filter="

    # ページの取得
    time.sleep(3)
    page = requests.get(urlName)
    page.encoding = page.apparent_encoding

    # BeautifulSoupオブジェクトの生成
    soup = BeautifulSoup(page.content, 'html.parser')
    #pprint(soup)

    # 商品ブロックをすべて検索
    product_blocks = soup.find_all('div', class_='l-product__wrap')

    for block in product_blocks:
        # 商品名の取得
        title_tag = block.find('h3', class_='l-product__wrap--middle__productTitle')
        if title_tag:
            item_names.append(title_tag.get_text(strip=True))
        else:
            item_names.append("")

        # 商品ラベルの取得
        label_tag = block.find('div', class_='l-product__wrap--top__labelTop__leftArea')
        if label_tag:
            label.append(label_tag.get_text(strip=True))
        else:
            label.append("")

        # 商品カテゴリの取得
        category_tag = block.find('div', class_='l-product__wrap--top__labelBottom')
        if category_tag:
            category.append(category_tag.get_text(strip=True))
        else:
            category.append("")
            
```

```python
# リストをデータフレームの列に変換
df = pd.DataFrame({'item': item_names, 'label':label, 'category': category})

display(df)
```
